```python
import matplotlib.pyplot as plt
from skimage import io
img = io.imread('baboon.png')
```

```python
print(img[0])
print(img[0,0])
print(img[0,0,0])
```

```python
plt.figure(figsize=(4, 4))
plt.imshow(img, cmap='gray', interpolation='nearest')
plt.axis('on')
plt.tight_layout()
plt.show()
```

```python
import numpy as np
test_image = img.copy()
lx, ly, lz = test_image.shape
X, Y = np.ogrid[0:lx, 0:ly]
mask = (X - lx / 2) ** 2 + (Y - ly / 2) ** 2 > lx * ly / 8;
test_image[mask]=0;
io.imshow(test_image)
io.show()
# vignette image
```

```python
camera_multiply = img*3
io.imshow(camera_multiply)
io.show()
# inverse color not a negative color  inverse 
```

```python
from skimage.transform import resize, rotate, swirl
resized = resize(img, (350, 300), mode='constant', anti_aliasing=True)
io.imshow(resized)
io.show()
# AA or Anti - Aliasing use to increase sharpness for image rendering but may decrease the pc performance
```

```python
from skimage.transform import resize, rotate, swirl
resized = resize(img, (50, 50), mode='constant', anti_aliasing=True)
# resize to 50 * 50 the quality of image is blurrly 
io.imshow(resized)
io.show()
# AA or Anti - Aliasing use to increase sharpness for image rendering but may decrease the pc performance
```

```python
#Rotate image by a certain angle around its center.
#Rotation angle in degrees in counter-clockwise direction.
rotate = rotate(img, 90)
io.imshow(rotate)
io.show()
```

```python
swirled = swirl(img, rotation=0, strength=150, radius=180, mode='constant')
io.imshow(swirled)
io.show()
# swirled from center 
```

```python
from skimage import filters
blurred_img = filters.gaussian(img, sigma=5, multichannel=True)
io.imshow(blurred_img)
io.show()
# using filter gaussian blur 
```

```python
alpha = 0.15
shapened_img = img + alpha * (img - blurred_img)
io.imshow(shapened_img.astype(int))
io.show()
io.imshow(img)
io.show()
# increase by calculate from base image and blurrly image 
```
